```python
# === 🔹 PHASE 4: Load Iris Dataset for Logical Reasoning 🔹 ===
from sklearn.datasets import load_iris

# Load dataset
iris = load_iris()
X_iris = iris.data  # Features: Sepal & Petal dimensions
y_iris = iris.target  # Labels: 0 = Setosa, 1 = Versicolor, 2 = Virginica

# Map labels to flower names
target_names = iris.target_names

# Show sample
print("Sample Features (first row):", X_iris[0])
print("Target Name:", target_names[y_iris[0]])
```

```python
# Train/Test split
X_train_iris, X_test_iris, y_train_iris, y_test_iris = train_test_split(
    X_iris, y_iris, test_size=0.2, random_state=42)

print("✅ Iris data split complete: Training on", len(X_train_iris), "samples.")
```

```python
# Simulate neuron firing based on cosine similarity to "goal vector" (flower class vector)

# Pick goal vector for 'setosa' (0)
goal_vector = np.mean(X_train_iris[y_train_iris == 0], axis=0)

# Cosine similarity function
def get_similarity(v1, v2):
    return cosine_similarity([v1], [v2])[0][0]

# Run simulation: match training examples to goal
firing_results = []

for i in range(len(X_train_iris)):
    input_vec = X_train_iris[i]
    similarity = get_similarity(input_vec, goal_vector)

    # Trigger neurons based on similarity
    cell_idx = i % NUM_CELLS
    neuron_idx = i % NEURONS_PER_CELL
    neuron_outputs[cell_idx][neuron_idx] = similarity

    firing_results.append((i, similarity, target_names[y_train_iris[i]]))

print("✅ Iris inputs processed & matched to neuron firing patterns.")
```

```python
# Convert to DataFrame for graph
firing_df = pd.DataFrame(firing_results, columns=["Index", "Similarity", "Class"])

# Plot similarity scores
plt.figure(figsize=(8,5))
sns.boxplot(data=firing_df, x="Class", y="Similarity", palette="coolwarm")
plt.title("Neuron Firing Similarity by Iris Class")
plt.show()
```

```python
# === 🔹 Train ML Model: Decision Tree Classifier ===
from sklearn.tree import DecisionTreeClassifier

# Train
ml_model = DecisionTreeClassifier()
ml_model.fit(X_train_iris, y_train_iris)

# Predict
ml_preds = ml_model.predict(X_test_iris)

# Accuracy
ml_acc = accuracy_score(y_test_iris, ml_preds)
print(f"✅ ML Model Accuracy: {ml_acc:.4f}")
```

```python
# === 🔹 Train DL Model (Keras FFN) ===

# One-hot encode target
y_train_dl = to_categorical(y_train_iris, num_classes=3)
y_test_dl = to_categorical(y_test_iris, num_classes=3)

# Define DL model
dl_model = Sequential([
    Dense(16, input_shape=(4,), activation='relu'),
    Dropout(0.1),
    Dense(12, activation='relu'),
    Dense(3, activation='softmax')
])

dl_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train
dl_model.fit(X_train_iris, y_train_dl, epochs=20, batch_size=8, verbose=0)

# Evaluate
dl_loss, dl_acc = dl_model.evaluate(X_test_iris, y_test_dl, verbose=0)
print(f"✅ DL Model Accuracy: {dl_acc:.4f}")
```

```python
fused_preds = []
alpha, beta, gamma = 0.2, 0.4, 0.4  # Adjust weights: more trust to ML/DL

for i in range(len(X_test_iris)):
    input_vec = X_test_iris[i]

    # --- AGI Similarity → Fake Probabilities (3-class soft logic) ---
    sim = get_similarity(input_vec, goal_vector)
    agi_probs = [1 - sim, sim / 2, sim / 2]  # distribute across 3 classes

    # --- ML Probabilities ---
    ml_prob = np.zeros(3)
    ml_prob[ml_preds[i]] = 1

    # --- DL Probabilities ---
    dl_prob = dl_model.predict(np.array([input_vec]), verbose=0)[0]

    # --- Blended Output (Softmax Fusion) ---
    final_probs = (
        alpha * np.array(agi_probs) +
        beta * np.array(ml_prob) +
        gamma * np.array(dl_prob)
    )

    fused_class = np.argmax(final_probs)
    fused_preds.append(fused_class)

# Accuracy
fused_acc = accuracy_score(y_test_iris, fused_preds)
print(f"✅ FIXED Fusion Accuracy: {fused_acc:.4f}")
```

```python
# === 🔹 Dharma Ethics Classifier (Sattva / Rajas / Tamas) ===

def classify_dharma(emotion_text):
    # Basic emotion → dharma map
    sattva_keywords = ['joy', 'love', 'peace', 'happy', 'gratitude']
    rajas_keywords  = ['surprise', 'curious', 'excited']
    tamas_keywords  = ['anger', 'fear', 'sad', 'anxious', 'helpless', 'frustrated']

    text = emotion_text.lower()

    for word in sattva_keywords:
        if word in text:
            return 'sattva'
    for word in rajas_keywords:
        if word in text:
            return 'rajas'
    for word in tamas_keywords:
        if word in text:
            return 'tamas'

    return 'rajas'  # default to neutral if unknown
```

```python
# === 🔹 Memory System: Short-Term + Long-Term ===

# Short-Term Memory (fixed stack of last N actions)
SHORT_TERM_MEMORY = []

# Long-Term Memory (persistent structured log)
LONG_TERM_MEMORY = {
    'emotion': [],
    'iris_class': [],
    'neuron_output': [],
    'dharma': [],
    'correct': []
}
```

```python
# Add to memory (for last 20 only in short term)
if len(SHORT_TERM_MEMORY) >= 20:
    SHORT_TERM_MEMORY.pop(0)  # remove oldest
SHORT_TERM_MEMORY.append({
    'input': input_vec,
    'agi_sim': sim,
    'ml': ml_raw,
    'dl': dl_raw,
    'fused': fused_class,
    'true': y_test_iris[i]
})

# Add to long-term
LONG_TERM_MEMORY['iris_class'].append(y_test_iris[i])
LONG_TERM_MEMORY['neuron_output'].append(sim)
LONG_TERM_MEMORY['emotion'].append('unknown')
LONG_TERM_MEMORY['dharma'].append('neutral')
LONG_TERM_MEMORY['correct'].append(int(fused_class == y_test_iris[i]))
```

```python
# Self-correct neuron weight if prediction was wrong
def replay_and_correct():
    corrections = 0
    for mem in SHORT_TERM_MEMORY:
        if mem['fused'] != mem['true']:
            # simulate weight adjustment on neurons
            adjustment = 0.05 if mem['agi_sim'] < 0.5 else -0.05
            cell = random.randint(0, NUM_CELLS - 1)
            neuron = random.randint(0, NEURONS_PER_CELL - 1)
            neuron_weights[cell][neuron] += adjustment
            corrections += 1
    print(f"🔁 Replay triggered: {corrections} neurons adjusted.")

# Trigger self-correction
replay_and_correct()
```

```python
# Accuracy from memory logs
mem_acc = sum(LONG_TERM_MEMORY['correct']) / len(LONG_TERM_MEMORY['correct'])
print(f"📊 Long-Term Memory Accuracy: {mem_acc:.4f}")
```

```python
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

# Load CSV
df = pd.read_csv("sentiment140.csv", encoding='latin1')
df = df[['tweet', 'label']]
df.columns = ['text', 'sentiment']

# Map sentiment labels to text
label_map = {0: 'negative', 2: 'neutral', 4: 'positive'}
df['sentiment'] = df['sentiment'].map(label_map)

# Drop NaNs and clean text
df = df.dropna()
df = df[df['text'].apply(lambda x: isinstance(x, str))]

def clean_tweet(text):
    text = re.sub(r"http\S+|[^a-zA-Z\s]", "", text)
    return text.lower().strip()

df['clean_text'] = df['text'].apply(clean_tweet)

# Vectorize
vectorizer = TfidfVectorizer(max_features=300)
X_tweet = vectorizer.fit_transform(df['clean_text']).toarray()

# Encode sentiment
le_tweet = LabelEncoder()
y_tweet = le_tweet.fit_transform(df['sentiment'])

# Split
X_train_tweet, X_test_tweet, y_train_tweet, y_test_tweet = train_test_split(
    X_tweet, y_tweet, test_size=0.2, random_state=42
)

print("✅ Loaded and preprocessed Sentiment140 dataset!")
print("Classes:", le_tweet.classes_)
```

```python
# === 🔹 Phase X-C: Train NeuronMLX AGI on Sentiment140 Tweet Data ===

import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# AGI Brain Config
NUM_CELLS = 30000
NEURONS_PER_CELL = 96
VECTOR_DIM = 300  # TF-IDF size

# Reinit AGI Brain
neuron_weights = np.random.normal(0.5, 0.1, size=(NUM_CELLS, NEURONS_PER_CELL, VECTOR_DIM))
neuron_modes = np.random.choice(['excite', 'inhibit', 'calm', 'hold'], size=(NUM_CELLS, NEURONS_PER_CELL))
ethics_tags = np.random.choice(['sattva', 'rajas', 'tamas'], size=(NUM_CELLS, NEURONS_PER_CELL))

# Ethics Mapping Based on Sentiment
sentiment_ethics_map = {
    'positive': 'sattva',
    'neutral': 'rajas',
    'negative': 'tamas'
}

# === 🧠 Train AGI Brain ===
for i in range(min(len(X_train_tweet), NUM_CELLS * NEURONS_PER_CELL)):
    cell_idx = i % NUM_CELLS
    neuron_idx = i % NEURONS_PER_CELL

    vec = X_train_tweet[i]
    sentiment_idx = y_train_tweet[i]
    sentiment_label = le_tweet.classes_[sentiment_idx]

    # Ethics
    ethics_tags[cell_idx][neuron_idx] = sentiment_ethics_map.get(sentiment_label, 'rajas')

    # Mode logic
    if sentiment_label == 'negative':
        neuron_modes[cell_idx][neuron_idx] = 'inhibit'
    elif sentiment_label == 'positive':
        neuron_modes[cell_idx][neuron_idx] = 'calm'
    else:
        neuron_modes[cell_idx][neuron_idx] = 'excite'

    # Train weights
    neuron_weights[cell_idx][neuron_idx] += vec * 0.01  # tiny learning step

print(f"✅ AGI Brain Training Complete on Sentiment140 ({NUM_CELLS * NEURONS_PER_CELL:,} neurons)")
```

```python
# === 🔍 Test neuron output on 1 tweet ===
test_idx = 42
test_vec = X_test_tweet[test_idx]
true_label = le_tweet.classes_[y_test_tweet[test_idx]]

cell = test_idx % NUM_CELLS
neuron = test_idx % NEURONS_PER_CELL

similarity = cosine_similarity([test_vec], [neuron_weights[cell][neuron]])[0][0]

print(f"🧠 Neuron [{cell}][{neuron}] similarity: {similarity:.4f}")
print(f"📄 True sentiment: {true_label}")
print(f"🧭 Ethics: {ethics_tags[cell][neuron]}")
print(f"⚙️ Mode: {neuron_modes[cell][neuron]}")
```

```python
TRAINING_LOOPS = 3  # You can try 5 or even 10 on TPU

for loop in range(TRAINING_LOOPS):
    print(f"🔁 Loop {loop+1}/{TRAINING_LOOPS}")
    for i in range(min(len(X_train_tweet), NUM_CELLS * NEURONS_PER_CELL)):
        cell_idx = i % NUM_CELLS
        neuron_idx = i % NEURONS_PER_CELL

        vec = X_train_tweet[i]
        sentiment_idx = y_train_tweet[i]
        sentiment_label = le_tweet.classes_[sentiment_idx]

        # Ethics
        ethics_tags[cell_idx][neuron_idx] = sentiment_ethics_map.get(sentiment_label, 'rajas')

        # Mode logic
        if sentiment_label == 'negative':
            neuron_modes[cell_idx][neuron_idx] = 'inhibit'
        elif sentiment_label == 'positive':
            neuron_modes[cell_idx][neuron_idx] = 'calm'
        else:
            neuron_modes[cell_idx][neuron_idx] = 'excite'

        # ✅ Fusion: Learn more with decay and control
        update = vec * (0.01 / (loop + 1))  # Gradual decrease to stabilize
        neuron_weights[cell_idx][neuron_idx] += update
```

```python
# 🔍 Test overall AGI accuracy
correct = 0
total = len(X_test_tweet)

for i in range(total):
    vec = X_test_tweet[i]
    label_idx = y_test_tweet[i]

    cell = i % NUM_CELLS
    neuron = i % NEURONS_PER_CELL

    similarity = cosine_similarity([vec], [neuron_weights[cell][neuron]])[0][0]

    # Basic logic: if similarity high → match ethics tag with label
    predicted_sentiment = ethics_tags[cell][neuron]
    actual_sentiment = sentiment_ethics_map[le_tweet.classes_[label_idx]]

    if predicted_sentiment == actual_sentiment:
        correct += 1

accuracy = correct / total
print(f"🎯 AGI Ethics Accuracy: {accuracy:.4f}")
```

```python
import pandas as pd
import json
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

# Load the DROP dataset
with open("drop_dataset_train.json", "r") as f:
    drop_data = json.load(f)

# Parse into structured QA data
qa_rows = []
for passage_id, passage_data in drop_data.items():
    passage = passage_data.get('passage', '')
    for qa in passage_data.get('qa_pairs', []):
        question = qa.get('question', '')
        answers = qa.get('answer', {}).get('spans', [])
        if answers:
            qa_rows.append({
                'question': question,
                'passage': passage,
                'answer': answers[0]  # Use the first span answer
            })

# Create DataFrame
df = pd.DataFrame(qa_rows)
print(f"✅ Loaded {len(df)} QA pairs from DROP dataset")

# Clean & Combine Input
def clean_text(text):
    return re.sub(r'[^a-zA-Z0-9\s]', '', text.lower().strip())

df['input'] = df['question'] + ' ' + df['passage']
df['clean_input'] = df['input'].apply(clean_text)

# Vectorize using TF-IDF
vectorizer = TfidfVectorizer(max_features=300)
X_drop = vectorizer.fit_transform(df['clean_input']).toarray()

# Encode answer (for now just a basic label to simulate output)
le_drop = LabelEncoder()
y_drop = le_drop.fit_transform(df['answer'])

# Train/Test split
X_train_drop, X_test_drop, y_train_drop, y_test_drop = train_test_split(
    X_drop, y_drop, test_size=0.2, random_state=42
)

print("✅ DROP dataset is ready for AGI training!")
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# === AGI NeuronMLX Brain Config ===
NUM_CELLS = 30000
NEURONS_PER_CELL = 96
VECTOR_DIM = 300

# Reinitialize Brain
neuron_weights = np.random.normal(0.5, 0.1, size=(NUM_CELLS, NEURONS_PER_CELL, VECTOR_DIM))
neuron_modes = np.random.choice(['excite', 'inhibit', 'calm', 'hold'], size=(NUM_CELLS, NEURONS_PER_CELL))
ethics_tags = np.random.choice(['sattva', 'rajas', 'tamas'], size=(NUM_CELLS, NEURONS_PER_CELL))

# Ethics logic linked to answer index
answer_ethics_map = ['sattva', 'rajas', 'tamas']

# === 🔁 Deep Reasoning Loops ===
TRAINING_LOOPS = 25  # Deep cognition

for loop in range(TRAINING_LOOPS):
    print(f"🔁 Training Loop {loop+1}/{TRAINING_LOOPS}")
    for i in range(min(len(X_train_drop), NUM_CELLS * NEURONS_PER_CELL)):
        cell_idx = i % NUM_CELLS
        neuron_idx = i % NEURONS_PER_CELL

        vec = X_train_drop[i]
        answer_idx = y_train_drop[i]
        ethics = answer_ethics_map[answer_idx % 3]

        # Ethics Tagging
        ethics_tags[cell_idx][neuron_idx] = ethics

        # Mode Assignment
        if ethics == 'sattva':
            neuron_modes[cell_idx][neuron_idx] = 'calm'
        elif ethics == 'rajas':
            neuron_modes[cell_idx][neuron_idx] = 'excite'
        else:
            neuron_modes[cell_idx][neuron_idx] = 'inhibit'

        # Smart update with decay (strong early, soft later)
        update = vec * (0.01 / (loop + 1))
        neuron_weights[cell_idx][neuron_idx] += update

print("✅ ✅ ✅ AGI Training Complete with 25 Deep Loops!")
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import random

# === Config ===
NUM_CELLS = 30000
NEURONS_PER_CELL = 96
VECTOR_DIM = 300
TRAINING_LOOPS = 10
NEUROGENESIS_THRESHOLD = 0.2  # Trigger new neurons if similarity is too low

# === Initialize Brain ===
neuron_weights = np.random.normal(0.5, 0.1, (NUM_CELLS, NEURONS_PER_CELL, VECTOR_DIM))
ethics_tags = np.random.choice(['sattva', 'rajas', 'tamas'], size=(NUM_CELLS, NEURONS_PER_CELL))
neuron_modes = np.random.choice(['excite', 'inhibit', 'calm', 'hold'], size=(NUM_CELLS, NEURONS_PER_CELL))
answer_ethics_map = ['sattva', 'rajas', 'tamas']

# === PHASE 5: Neurogenesis ===
def neurogenesis(vec, cell, neuron):
    similarity = cosine_similarity([vec], [neuron_weights[cell][neuron]])[0][0]
    if similarity < NEUROGENESIS_THRESHOLD:
        new_weight = vec + np.random.normal(0, 0.05, VECTOR_DIM)
        neuron_weights[cell][neuron] = new_weight
        ethics_tags[cell][neuron] = random.choice(answer_ethics_map)
        neuron_modes[cell][neuron] = random.choice(['excite', 'calm'])
        return True
    return False

# === PHASE 6: Meta-Learning (Transfer from similar vecs) ===
def meta_learn(vec):
    idx = np.random.randint(len(X_train_drop))
    transferred_vec = X_train_drop[idx]
    return (vec + transferred_vec) / 2

# === PHASE 7: Self-Modifying Neurons ===
def self_modify(cell, neuron, correct):
    if not correct:
        neuron_weights[cell][neuron] *= 0.95
        ethics_tags[cell][neuron] = random.choice(answer_ethics_map)
        neuron_modes[cell][neuron] = 'hold'

# === PHASE 8: Recursive Reasoning (2-step loop) ===
def recursive_reason(vec, steps=2):
    for _ in range(steps):
        vec = vec * np.mean(vec) + 0.1  # Simulated recursive update
    return vec

# === PHASE 9: Goal Engine ===
def solve_goal(vec, goal_type='match_sattva'):
    if goal_type == 'match_sattva':
        return vec * 1.05 + 0.02  # Slight push toward ethical vector
    return vec

# === PHASE 10: Tool Use Simulation ===
def tool_use(vec):
    tool_result = np.sin(vec) * 0.1  # Fake calculation simulating external tool
    return vec + tool_result

# === FULL TRAINING LOOP ===
for loop in range(TRAINING_LOOPS):
    print(f"🔁 Loop {loop+1}/{TRAINING_LOOPS}")
    for i in range(min(len(X_train_drop), NUM_CELLS * NEURONS_PER_CELL)):
        cell = i % NUM_CELLS
        neuron = i % NEURONS_PER_CELL

        vec = X_train_drop[i]
        label = y_train_drop[i]
        ethics = answer_ethics_map[label % 3]

        # Phase 6: meta-learn, 8: recursive reason, 9: solve goal
        vec = meta_learn(vec)
        vec = recursive_reason(vec)
        vec = solve_goal(vec)

        # Phase 10: tool simulation
        vec = tool_use(vec)

        # Phase 5: neurogenesis
        neurogenesis(vec, cell, neuron)

        # Ethics & mode tagging
        ethics_tags[cell][neuron] = ethics
        neuron_modes[cell][neuron] = 'excite' if ethics == 'rajas' else 'calm'

        # Weight update
        neuron_weights[cell][neuron] += vec * 0.005

print("✅ PHASE 5–10 Training Complete")

# === TESTING WITH SELF-MODIFYING CHECK ===
correct = 0
for i in range(len(X_test_drop)):
    vec = X_test_drop[i]
    true_idx = y_test_drop[i]
    true_ethics = answer_ethics_map[true_idx % 3]

    cell = i % NUM_CELLS
    neuron = i % NEURONS_PER_CELL

    similarity = cosine_similarity([vec], [neuron_weights[cell][neuron]])[0][0]
    predicted_ethics = ethics_tags[cell][neuron]
    is_correct = predicted_ethics == true_ethics

    # Phase 7: self-modification if incorrect
    self_modify(cell, neuron, is_correct)

    if is_correct:
        correct += 1

    if i % 1000 == 0:
        print(f"🧠 Neuron [{cell}][{neuron}] | Sim: {similarity:.4f} | True: {true_ethics} | Pred: {predicted_ethics} | Mode: {neuron_modes[cell][neuron]}")

print(f"\n🎯 FINAL AGI Accuracy (Post Phase10): {correct / len(X_test_drop):.4f} ✅")
```

```python
import numpy as np
import torch
import torch.nn as nn
from sklearn.metrics.pairwise import cosine_similarity
from collections import Counter

# ====== CONFIG ======
NUM_CELLS = 30000
NEURONS_PER_CELL = 96
TRAINING_LOOPS = 10
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ====== GET VECTOR SIZE DYNAMICALLY ======
VECTOR_DIM = X_train_drop.shape[1]

# ====== NEURON STORAGE (CPU - symbolic part) ======
neuron_weights = np.random.rand(NUM_CELLS, NEURONS_PER_CELL, VECTOR_DIM)
ethics_tags = np.random.choice(['sattva', 'rajas', 'tamas'], (NUM_CELLS, NEURONS_PER_CELL))
neuron_modes = np.random.choice(['excite', 'inhibit', 'calm', 'hold'], (NUM_CELLS, NEURONS_PER_CELL))

# ====== SIMPLE MLP REFINER (Backprop model) ======
class AGIRefiner(nn.Module):
    def __init__(self, input_dim=300, output_dim=3):
        super().__init__()
        self.layer1 = nn.Linear(input_dim, 128)
        self.relu = nn.ReLU()
        self.out = nn.Linear(128, output_dim)

    def forward(self, x):
        x = self.relu(self.layer1(x))
        return self.out(x)

model = AGIRefiner(input_dim=VECTOR_DIM, output_dim=3).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# ====== CONVERT TO GPU TENSORS ======
X_tensor = torch.tensor(X_train_drop, dtype=torch.float32).to(device)
y_tensor = torch.tensor(y_ethics_encoded, dtype=torch.long).to(device)
# === AUTO ALIGN X and y ===
min_len = min(len(X_train_drop), len(y_ethics_encoded))
X_train_drop = X_train_drop[:min_len]
y_ethics_encoded = y_ethics_encoded[:min_len]

# === Convert to GPU tensors ===
X_tensor = torch.tensor(X_train_drop, dtype=torch.float32).to(device)
y_tensor = torch.tensor(y_ethics_encoded, dtype=torch.long).to(device)

# ====== GPU TRAINING + NEURON FUSION ======
for epoch in range(TRAINING_LOOPS):
    print(f"🔁 Loop {epoch+1}/{TRAINING_LOOPS}")
    model.train()
    optimizer.zero_grad()
    outputs = model(X_tensor)
    loss = criterion(outputs, y_tensor)
    loss.backward()
    optimizer.step()

    # === Symbolic vector fusion on CPU ===
    with torch.no_grad():
        for i in range(min(len(X_train_drop), NUM_CELLS * NEURONS_PER_CELL)):
            cell = i % NUM_CELLS
            neuron = i % NEURONS_PER_CELL
            vec = X_train_drop[i]
            neuron_weights[cell][neuron] += vec * 0.001  # Small fusion boost

print("✅ PHASE 5–10 GPU Training Complete and Synced with Neuron Logic")
```

```python
import numpy as np
import plotly.graph_objects as go
import random

# === CONFIG ===
NUM_CELLS = 50
NEURONS_PER_CELL = 40
TOTAL_NEURONS = NUM_CELLS * NEURONS_PER_CELL
LOOPS = 15

# === Generate Neuron Positions Around Cells ===
np.random.seed(42)
cell_centers = np.random.rand(NUM_CELLS, 3) * 30
neuron_positions = []

for center in cell_centers:
    cluster = center + np.random.normal(scale=0.8, size=(NEURONS_PER_CELL, 3))
    neuron_positions.append(cluster)

neuron_positions = np.concatenate(neuron_positions, axis=0)

# === Assign Neuron Base Colors (faded premium palette) ===
base_colors = ['deepskyblue', 'violet', 'gold', 'tomato', 'lime']
neuron_colors = [random.choice(base_colors) for _ in range(TOTAL_NEURONS)]

# === Create Firing Animations ===
frames = []
for loop in range(LOOPS):
    # Randomly select firing neurons (about 5% each loop)
    firing_indices = np.random.choice(TOTAL_NEURONS, int(TOTAL_NEURONS * 0.05), replace=False)

    colors = []
    sizes = []
    for i in range(TOTAL_NEURONS):
        if i in firing_indices:
            colors.append('red')
            sizes.append(6)  # Firing neurons bigger
        else:
            colors.append(neuron_colors[i])
            sizes.append(3)

    frames.append(go.Frame(
        data=[go.Scatter3d(
            x=neuron_positions[:, 0],
            y=neuron_positions[:, 1],
            z=neuron_positions[:, 2],
            mode='markers',
            marker=dict(
                size=sizes,
                color=colors,
                opacity=0.88,
                symbol='circle',
                line=dict(width=1, color='white')
            )
        )],
        name=f"Loop {loop+1}"
    ))

# === Base Trace ===
init_trace = go.Scatter3d(
    x=neuron_positions[:, 0],
    y=neuron_positions[:, 1],
    z=neuron_positions[:, 2],
    mode='markers',
    marker=dict(
        size=3,
        color=neuron_colors,
        opacity=0.88,
        line=dict(width=1, color='white')
    ),
    name='Neurons'
)

# === Cell Trace ===
cell_trace = go.Scatter3d(
    x=cell_centers[:, 0],
    y=cell_centers[:, 1],
    z=cell_centers[:, 2],
    mode='markers',
    marker=dict(
        size=8,
        color='white',
        opacity=1,
        symbol='diamond',
        line=dict(width=2, color='black')
    ),
    name='Cells'
)

# === Figure Layout ===
fig = go.Figure(
    data=[init_trace, cell_trace],
    frames=frames,
    layout=go.Layout(
        title="🧠 NeuronMLX AGI – Premium Live Neuron Firing",
        scene=dict(
            xaxis=dict(title='X', showgrid=False),
            yaxis=dict(title='Y', showgrid=False),
            zaxis=dict(title='Z', showgrid=False),
            bgcolor='white'
        ),
        paper_bgcolor='white',
        updatemenus=[dict(
            type="buttons",
            showactive=False,
            buttons=[dict(label="▶️ Play", method="animate", args=[None, {"frame": {"duration": 500}, "fromcurrent": True}])]
        )],
        margin=dict(l=0, r=0, t=40, b=0)
    )
)

fig.show()
```
